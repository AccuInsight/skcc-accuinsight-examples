### Decomposition

```python
components = seasonal_decompose(sample_df['Sales'], model='additive', freq=30)
ts = (sample_df.Sales.to_frame('Original')
      .assign(Trend=components.trend)
      .assign(Seasonality=components.seasonal)
      .assign(Residual=components.resid))
_ = ts.plot(subplots=True, figsize=(14, 8))
```

- 전체적인 추세는 없음, 12월(크리스마스 시즌)에 뚜렷한 peak

### ADF Test
- 시계열의 정상성(staionary)를 검정
  - 귀무가설: 단위근을 가진다(비정상성, 안정적이지 않다)
  - 대립가설: 안정적이다.(정상성)
  - 귀무가설을 기각한다(=정상 시계열이 아니라)면, 추가적인 차분을 고려

```python
result = list(adfuller(sample_df['Sales']))
print('ADF Statistic: {}'.format(result[0]))
print('p-value: {}'.format(result[1]))
print('Critical Values:')
for key, value in result[4].items():
    print('\t{}: {}'.format(key, value))
```

## Road Data

```python
file_name = os.path.join(data_processed_dir, 'rossman_sample_final.csv')
target_df = pd.read_csv(file_name, parse_dates=True, index_col=0)
```

## Library for ML

```python
from sklearn.preprocessing import OneHotEncoder, PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
```

## Data Prepare
ML Modeling 위한 데이터 준비 
- pd.DataFrame -> np.array로 변환
- one-hot encoding
- Polynomial(degree=2), Interaction 변수 추가
- Sales Lag1~5 feature 추가 

```python
target_df.dtypes
```

```python
y = target_df['Sales'].values.reshape(-1,1)
X = np.hstack([target_df['DayOfWeek'].values.reshape(-1,1),
               target_df['Open'].values.reshape(-1,1), 
               target_df['Promo'].values.reshape(-1,1), 
               target_df['StateHoliday'].values.reshape(-1,1), 
               target_df['SchoolHoliday'].values.reshape(-1,1)])
```

```python
enc = OneHotEncoder() 
X_onehot = enc.fit_transform(X).toarray()
X_names_onehot = enc.get_feature_names(['DayOfWeek', 'Open', 'Promo', 'StateHoliday', 'SchoolHoliday'])
print(X_onehot.shape)
```

```python
poly_transformer = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_onehot_poly = poly_transformer.fit_transform(X_onehot)
X_names_onehot_poly = poly_transformer.get_feature_names(X_names_onehot)

print(X_onehot_poly.shape)
print(X_names_onehot_poly)
```

lag1~5를 Feature에 추가하고 Test

```python
target_df['lag1_sales'] = target_df['Sales'].shift(1)
target_df['lag2_sales'] = target_df['Sales'].shift(2)
target_df['lag3_sales'] = target_df['Sales'].shift(3)
target_df['lag4_sales'] = target_df['Sales'].shift(4)
target_df['lag5_sales'] = target_df['Sales'].shift(5)

target_df = target_df.fillna(0)

X_lag5 = np.hstack([target_df['lag1_sales'].values.reshape(-1,1),
                                 target_df['lag2_sales'].values.reshape(-1,1),
                                  target_df['lag3_sales'].values.reshape(-1,1),
                                   target_df['lag4_sales'].values.reshape(-1,1),
                                    target_df['lag5_sales'].values.reshape(-1,1)])

X_onehot_poly_lag5 = np.hstack((X_lag5, X_onehot_poly))
X_names_onehot_poly_lag5 = ['lag1','lag2','lag3','lag4','lag5']+X_names_onehot_poly

print(X_onehot_poly_lag5.shape)
print(X_names_onehot_poly_lag5[:10])
```

```python
n_train = np.round(len(target_df) * 0.7).astype('int')

# start_date:  2013-01-01 00:00:00
# end_date:  2015-07-31 00:00:00
date_ticks = pd.date_range(start="2013-01-01", end="2015-07-31", freq='D')

def eval_on_features(features, target, regressor):
    x_train, x_test = features[:n_train], features[n_train:]
    y_train, y_test = target[:n_train], target[n_train:]
    
    regressor.fit(x_train, y_train)
    
    y_pred = regressor.predict(x_test)
    y_pred_train = regressor.predict(x_train)
    
    print("Train MAE: {:.2f}".format(mean_absolute_error(y_train, y_pred_train)))
    print("Test MAE: {:.2f}".format(mean_absolute_error(y_test, y_pred)))
    # print("Train RMSE: {:.2f}".format(mean_squared_error(y_train, y_pred_train)))
    # print("Test RMSE: {:.2f}".format(mean_squared_error(y_test, y_pred))) # squared=True is default
    
    plt.figure(figsize=(15, 5))
    plt.xticks(date_ticks, rotation=90, ha='left')
    
    plt.plot(range(n_train), y_train, label="train")
    plt.plot(range(n_train, len(y_test) + n_train), y_test, '-', label="test")
    plt.plot(range(n_train), y_pred_train, '--', label="pred train")
    plt.plot(range(n_train, len(y_test) + n_train), y_pred, '--', label="pred test")
    plt.legend(loc=(1.01, 0))
```

## Linear Regression
- Input feature를 X, X_onehot, X_onehot_poly, X_onehot_poly_lag5로 변경해보면서 성능 확인

```python
eval_on_features(X, y, LinearRegression())
```

```python
eval_on_features(X_onehot, y, LinearRegression())
```

```python
eval_on_features(X_onehot_poly, y, LinearRegression())
```

```python
eval_on_features(X_onehot_poly_lag5, y, LinearRegression())
```

## Ridge Regression

```python
rlr = Ridge()
eval_on_features(X_onehot, y, rlr)
```

```python
eval_on_features(X_onehot_poly, y, rlr)
```

```python
eval_on_features(X_onehot_poly_lag5, y, rlr)
```

### Coefficient Check
- Train/Test 성능이 가장 좋은 마지막 모델(ridge, input = X_onehot_poly_lag5)의 변수별 coef 확인

```python
ridge_coef = np.array(rlr.coef_).reshape(rlr.coef_.shape[-1])
ridge_coef_df = pd.DataFrame(list(zip(X_names_onehot_poly_lag5, ridge_coef)), 
                             columns=['X_name', 'ridge_coef'])
threshhold = 0.5
high_ridge_coef = ridge_coef_df[ridge_coef_df['ridge_coef'] > threshhold]
high_ridge_coef = high_ridge_coef.sort_values(by=['ridge_coef'], ascending=False)
high_ridge_coef.head()
```

```python
plt.figure(figsize=(15,2))
plt.bar(range(high_ridge_coef.shape[0]), high_ridge_coef['ridge_coef'], color="r", align="center")
plt.xticks(range(high_ridge_coef.shape[0]), high_ridge_coef['X_name'], rotation=90)
plt.show()
```

- coef가 0.005보다 큰 것만 plotting

## RandomForest

```python
rfr = RandomForestRegressor(n_estimators=100, random_state=0)
eval_on_features(X_onehot, y, rfr)
```

```python
eval_on_features(X_onehot_poly, y, rfr)
```

```python
eval_on_features(X_onehot_poly_lag5, y, rfr)
```

### Variable Importance
- 가장 성능이 좋은 마지막 rfr모델(input = X_onehot_poly_lag5)의 변수 중요도 확인

```python
rfr_importance = rfr.feature_importances_
```

```python
importance_df = pd.DataFrame(list(zip(X_names_onehot_poly_lag5, rfr_importance)), columns=['X_name', 'importances'])
threshhold = 0.005
high_importances = importance_df[importance_df['importances']>threshhold]
high_importances = high_importances.sort_values(by=['importances'], ascending=False)
print(high_importances)
```

```python
plt.figure()
plt.title("Feature importances")
plt.bar(range(high_importances.shape[0]), high_importances['importances'], color="r", align="center")
plt.xticks(range(high_importances.shape[0]), high_importances['X_name'], rotation=90)
plt.xlim([-1, high_importances.shape[0]])
plt.show()
```
